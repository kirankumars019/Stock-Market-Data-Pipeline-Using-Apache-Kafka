```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
```

```python
#2-read data from csv to s3
s3 = S3FileSystem()
```

```python
#2-writting data from csv to s3
for count, i in enumerate(consumer):
    with s3.open("s3://kafka-stock-markets-kiran-6/stock_market_{}.json".format(count), 'w') as file:
        json.dump(i.value, file)
```

```python
#1-receiving data in consumer csv file from computer
# The consumer receives mixed / corrupted / invalid JSON messages -Using  Safe Deserializer
def safe_json(x):
    try:
        return loads(x.decode("utf-8"))
    except Exception:
        return {"invalid_json": x.decode("utf-8", errors="ignore")}
        
consumer = KafkaConsumer(
    'demo_testing_6',
    bootstrap_servers=['54.91.216.170:9092'],
    value_deserializer=safe_json,
    auto_offset_reset='earliest',
    enable_auto_commit=False
)
```

```python
#1-receiving data in consumer csv file from compute r
for msg in consumer:
    print("RAW VALUE:", msg.value)
```
